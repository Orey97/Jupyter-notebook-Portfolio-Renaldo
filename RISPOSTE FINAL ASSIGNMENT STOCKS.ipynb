```python
#Question 1: Use yfinance to Extract Stock Data
#Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function. 

tesla_data = yf.Ticker("TSLA")
tesla_df = tesla_data.history(period='max')

#reset index
tesla_df.reset_index(inplace = True)

#First 5 raws
print(tesla_df.head())
```

```python
#Question 2: Use Webscraping to Extract Tesla Revenue Data
#Display the last five rows of the tesla_revenue dataframe using the tail function.

#extract the data of revenue
tesla_revenue = tesla_data.financials.loc['Total Revenue']

#last 5 years total revenue 
print(tesla_revenue.tail())
```

```python
#Question 3: Use yfinance to Extract Stock Data
#Reset the index, save, and display the first five rows of the gme_data dataframe using the head function.

#extract data
gme_data = yf.Ticker("GME")
gme_df= gme_data.history(period = 'max')

#reset index
gme_df.reset_index(inplace = True)

#first 5 rows
print(gme_df.head())
```

```python
#Question 4: Use Webscraping to Extract GME Revenue Data
#Display the last five rows of the gme_revenue dataframe using the tail function

#extract the data of revenue
gme_revenue = gme_data.financials.loc['Total Revenue']

#last 5 years total revenue 
print(gme_revenue.tail())
```

```python
#Question 5: Plot Tesla Stock Graph
#Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph

import matplotlib.pyplot as plt

def make_graph (data, x_column, y_column, title, x_label, y_label):
    plt.plot(data[x_column], data[y_column], label=y_column, color='blue')
    plt.title(title, fontsize=16)  # Add graph title
    plt.grid(True)  # Add grid lines for better readability
    plt.legend()  # Add a legend
    plt.show()  # Display the grap


make_graph(
    data=tesla_df,
    x_column='Date',
    y_column='Close',
    title='Tesla Stock Closing Prices',
    x_label='Date',
    y_label='Closing Price (USD)'
)


```
